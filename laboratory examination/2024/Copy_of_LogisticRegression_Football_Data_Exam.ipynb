```python
from sklearn.model_selection import KFold
X=df.drop("Score",axis=1)
Y=df["Score"]
```

```python
from sklearn.linear_model import LogisticRegression  # Import Logistic Regression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score  # Import evaluation metrics

models={
        "Logistic Regression": LogisticRegression(max_iter=1000),
}
for model_name ,model in models.items():
    scores_f1 = []
    kf=KFold(n_splits=5,shuffle=False)
    for train_index , test_index in kf.split(X,Y):
      X_Train,X_Test = X.loc[train_index,:],X.loc[test_index,:]
      Y_Train,Y_Test = Y.loc[train_index],Y.iloc[test_index]
      model.fit(X_Train, Y_Train)  # Train the model on the training data
      Y_pred = model.predict(X_Test)
      scores_f1.append(f1_score(Y_Test, Y_pred, average='weighted'))  # Calculate F1-score
      print(f"{model_name} F1-Score: {np.mean(scores_f1):.4f}")

```

```python
logistic_importance = list(zip(X.columns, models["Logistic Regression"].coef_[0]))  # Get the coefficients
sorted_logistic_importance = sorted(logistic_importance, key=lambda x: abs(x[1]), reverse=True)  # Sort by absolute importance

# Extract sorted features and their coefficients
features, coefficients = zip(*sorted_logistic_importance)

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.barh(features, np.abs(coefficients), color='darkblue')  # Create a horizontal bar plot
plt.xlabel('Coefficient Value')
plt.ylabel('Features')
plt.title('Logistic Regression Feature Importance')
plt.gca().invert_yaxis()  # Invert y-axis to show the most important features at the top
plt.show()
```
